```python
import matplotlib.pyplot as plt
import matplotlib.collections as col
import pandas as pd
import random
import math
import numpy as np
```

```python
center=10
mx=20
artist_pd=pd.read_csv('artist_revolution_2020.csv')
artist_pd
base=7
x,y=[],[]
for i in range(artist_pd.shape[0]):
    dis=(85-artist_pd.loc[i].score)/base
    rd=random.uniform(0,2*math.pi)
    x.append(center+dis*math.sin(rd))
    y.append(center+dis*math.cos(rd))
```

```python
plt.figure(figsize=(18,18))
# plt.grid(True)

for i in range(15):
    plt.text(x[i]+0.1,y[i]+0.1,artist_pd.loc[i].artist_name+"("+str(artist_pd.loc[i].score)[:5]+")")
# plt.xlabel()
plt.xticks([i for i in range(mx+1)],[85-base*abs(i-center) for i in range(mx+1)])
plt.yticks([])
plt.xlabel("score")
plt.scatter(10,10,c='r',alpha=0.7)
plt.plot(x,y,'o',c='black')
# plt.plot([i for i in range(mx+1)],[center for i in range(mx+1)],marker='x',c='b')
for i in range(1,14):    
    circle1 = plt.Circle((10, 10), i, color='g',alpha=0.06)
    plt.gca().add_patch(circle1)
plt.savefig('raida.png')
```
