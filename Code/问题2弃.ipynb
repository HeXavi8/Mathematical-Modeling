```python
import pandas as pd
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
music_pd=pd.read_csv('full_music_data.csv')
music_pd=music_pd.drop(['artist_names','mode','key','explicit','popularity','year','release_date','song_title (censored)'],axis=1)
artist_pd=pd.read_csv('data_by_artist.csv')
artist_pd=artist_pd.drop(['mode','key','duration_ms','popularity','count'],axis=1)
```

```python
artist_attr={}
for i in range(artist_pd.shape[0]):
    artist_attr[artist_pd.loc[i]['artist_id']]=artist_pd.loc[i]['artist_name']
```

```python
for col in music_pd.columns:
    if(col=='artist_names' or col=='artists_id'): continue
    music_pd[col]=(music_pd[col]-music_pd[col].mean())/(music_pd[col].std())
```

```python
music_pd
```

```python
artist_music_attr={}
for i in range(music_pd.shape[0]):
    ids=music_pd.loc[i]['artists_id']
    ids=ids.lstrip('[').rstrip(']').split(",")
#     print(ids,len(ids),ids[0])
#     input()
    for artist_id in ids:
        artist_id=int(artist_id)
        if(artist_music_attr.setdefault(artist_id)==None):
            artist_music_attr[artist_id]={}
            for key,val in music_pd.loc[i].drop('artists_id').to_dict().items():
                artist_music_attr[artist_id][key]=val
            artist_music_attr[artist_id]['cnt']=1
        else:
            for key,val in music_pd.loc[i].drop('artists_id').to_dict().items():
                artist_music_attr[artist_id][key]+=val
            artist_music_attr[artist_id]['cnt']+=1
```

```python
# print(artist_music_attr)
with open('q2.txt','w+',encoding='utf-8') as f2:
    for key,val in artist_music_attr.items():
#         print(key,artist_attr[key])
        f2.write(artist_attr[key]+","+str(key))
        for k,v in val.items():
            if(k=="cnt"): continue
            f2.write(","+str(v/artist_music_attr[key]['cnt']))
        f2.write("\n")
```
